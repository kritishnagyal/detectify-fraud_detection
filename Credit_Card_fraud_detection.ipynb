Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

load the csv file

```python
cd=pd.read_csv('creditcard.csv')
cd
```

0 = Normal Transaction
1 = fraudulent transaction

```python
#separate the data 
l = cd[cd.Class == 0]
f = cd[cd.Class == 1]
```

```python
#know about the legal data
l.Amount.describe()
```

```python
#know about the fraud data
f.Amount.describe()
```

```python
# comparing values of both transactions
cd.groupby('Class').mean()
```

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions i.e. 492 correct and 492 fraud transactions

```python
l_sample = l.sample(n=492)
```

```python
cd_new = pd.concat([l_sample, f], axis=0)
cd_new
```

```python
cd_new.tail()
```

```python
#first 5 
cd_new.head()
```

```python
#new no. of correct and fraud transactions
cd_new['Class'].value_counts()
```

```python
cd_new.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
x = cd_new.drop(columns='Class', axis=1)
y = cd_new['Class']
```

```python
x
```

```python
y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#model selection
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
